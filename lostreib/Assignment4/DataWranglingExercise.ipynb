```python
#names each column
names = ['imdbID', 'title', 'year', 'score', 'votes', 'runtime', 'genres']
#says the columns are tab-separated, and column titles are those outlined above
data = pd.read_csv(url, delimiter='\t', names=names).dropna()
##not sure ##
print "Number of rows: %i" % data.shape[0]
data.head()  # print the first 5 rows
```

```python
#describes current string variable
dirty = '142 mins.'
#tells pandas to split the "dirty" string into a number and a text field, splits where there's a space
number, text = dirty.split(' ')
#describes the clean output as the variable now defined as number
clean = int(number)
# publish 
print number
```

```python
#create new named variable called clean_runtime, defined as the first string part (broken by a space) in the runtime column
clean_runtime = [float(r.split(' ')[0]) for r in data.runtime]
#data in runtime column is now the clean_runtime variable
data['runtime'] = clean_runtime
#print first 5 rows
data.head()
```

```python
#determine the unique genres
#setting the row using existing data
genres = set()
## defining the new data - what is "m" and "g" here??
for m in data.genres:
    genres.update(g for g in m.split('|'))
genres = sorted(genres)

#make a column for each genre
for genre in genres:
    data[genre] = [genre in movie.split('|') for movie in data.genres]
#print first five rows         
data.head()
```

```python
#prints the title characters minus the last seven in title column
data['title'] = [t[0:-7] for t in data.title]
#print first 5 rows
data.head()
```

```python
# print the summary statistics for those columns
data[['score', 'runtime', 'year', 'votes']].describe()
```
